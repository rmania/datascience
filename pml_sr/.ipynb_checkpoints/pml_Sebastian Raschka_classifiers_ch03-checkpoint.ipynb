```python
%matplotlib inline
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt
import warnings
from sklearn import datasets
from sklearn.cross_validation import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
iris = datasets.load_iris()
```

## Training a perceptron via sklearn

```python
# Iris dataset is a dict
iris.keys()
```

```python
# preprocessing , train, test split
X = pd.DataFrame(iris.data, columns = iris.feature_names).iloc[:,[2,3]].values
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= .3, random_state = 0)
```

```python
# scale the features
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

Having standardized the training data, we can now train our Perceptron. Most algorithms in sklearn already support multiclass classification by defauilt via the One-vs-rest (OvR) method, which allows us to feed the 3 flower classes to the Perceptron all at once.

here eta0 is equivalent to the learning rate eta from the handcrafted model. If this learning rate is too large, the algorithm will overshoot the global cost minimum, on the other hand, if it's too small, the algorithm will require more epochs until convergence, which can make the leanring slow.

```python
# fit the model
from sklearn.linear_model import Perceptron
ppn = Perceptron(n_iter=40, eta0=.1, random_state=0) # n_iter is number of Epochs (passes over the training set)
ppn.fit(X_train_std, y_train)
```

```python
# predict
y_pred = ppn.predict(X_test_std)
print ('Misclassified samples: %d' % (y_test != y_pred).sum())
```

```python
# large variety of metrics available via the metrics module
from sklearn.metrics import accuracy_score
print ('Accuracy: %.2f' % (accuracy_score(y_test, y_pred)))
```

function to plot the decision regions of our newly trained Perceptron , visualizing how well it separates the different flower samples.

- First we define no of colors and markers and create a color map via ListedColormap
- then we define the minumum/max values for the 2 features and use those feature vectors to create a pair of Grid arrays xx1, xx2 via the np.meshgrid function
-  we trained our maodel on 2 features, so we need to flatten the grid arrays and create a Matrix that has the same number of columns as the Iris training subset so that we can use the predict method to predict the class labels Z of the corresponding grid points
- after reshaping the predicted class labels Z into a grid w the same dims as xx1/xx2 cwe can now draw the contour plot via the contourf function, mapping the different decisio regions to different colors for each predicted class in the gris array

```python
def versiontuple(v):
    return tuple(map(int, (v.split("."))))


def plot_decision_regions(X, y, classifier, test_idx=None, resolution=0.02):

    # setup marker generator and color map
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])

    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.4, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1],
                    alpha=0.8, c=cmap(idx),
                    marker=markers[idx], label=cl)

    # highlight test samples
    if test_idx:
        # plot all samples
        if not versiontuple(np.__version__) >= versiontuple('1.9.0'):
            X_test, y_test = X[list(test_idx), :], y[list(test_idx)]
            warnings.warn('Please update to NumPy 1.9.0 or newer')
        else:
            X_test, y_test = X[test_idx, :], y[test_idx]

        plt.scatter(X_test[:, 0],
                    X_test[:, 1],
                    c='',
                    alpha=1.0,
                    linewidths=1,
                    marker='o',
                    s=55, label='test set')
```

```python
X_combined_std = np.vstack((X_train_std, X_test_std))
y_combined = np.hstack((y_train, y_test))

plot_decision_regions(X=X_combined_std, y=y_combined,
                      classifier=ppn, test_idx=range(105, 150))
plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
plt.tight_layout()
```

Perceptron never converges on datasets that are not perfectly linearly separable which is why we should be using more powerfull algorithms, linear classifiers that converge to a cost minumum even if the classes are not perfectly linearly separable.

## Modeling class probabilities via logistic regression 

despite its name this is a model for classification , NOT regression.

it is also a linear model for binary classificationthat can be extended to multiclass classififcation via the OvR technique.
The **logit** function takes input values in the range of 0 --> 1 and transforms these to values over the entire real number range, which we can use to express a linear relationship between feature values and the logg-odds. We are interested in knowing the probablility that a certain sample belongs to a particular class 1, given its features x., which is the inverse of the logit function, called the **logistic** function (sometimes also referred to as the sigmoid function due to its characheristic S-shape)

So it takes a real number value as input and transforms them into values in range [0,1] w. an intercept at $\phi = .5

- the output of the function is interpreted as the probability of a particular sample belonging to class 1, given its features x parameterized by the weights w.

```python
# let`s plot the sigmoid function for some values in range -7, 7 to see how it looks like
def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))

z = np.arange(-7, 7, 0.1)
phi_z = sigmoid(z)

plt.plot(z, phi_z)
plt.axvline(0.0, color='r')
plt.ylim(-0.1, 1.1)
plt.xlabel('z'); plt.ylabel('$\phi (z)$')

# y axis ticks and gridline
plt.yticks([0.0, 0.5, 1.0])
ax = plt.gca()
ax.yaxis.grid(True)
```

## Learning the weights of the logistic cost function



```python
def cost_1(z):
    return - np.log(sigmoid(z))


def cost_0(z):
    return - np.log(1 - sigmoid(z))

z = np.arange(-10, 10, 0.1)
phi_z = sigmoid(z)

c1 = [cost_1(x) for x in z]
plt.plot(phi_z, c1, label='J(w) if y=1')

c0 = [cost_0(x) for x in z]
plt.plot(phi_z, c0, linestyle='--', label='J(w) if y=0')

plt.ylim(0.0, 5.1)
plt.xlim([0, 1])
plt.xlabel('$\phi$(z)')
plt.ylabel('J(w)')
plt.legend(loc='best')
```

Plot illustrates the cost for the classification of a single-sample instance for different values o(z). We see that the cost approaches 0 (blue line)if we correctly predict that a sample belongs to class 1. However, if the prediction is wrong, the cost goes towards inf.

```python
# fit a LogisticRegression model
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(C = 1000, random_state=0) 
# C: Inverse of regularization strength; must be a pos. float: smaller values specify stronger regularization.

lr.fit(X_train_std, y_train)

plot_decision_regions(X_combined_std, y_combined, classifier=lr, test_idx=range(105,150))
plt.xlabel('petal_length [standardized]'); plt.ylabel('petal_width [standardized]')
plt.legend(loc= 'upper left')
```

## Tackling over-fitting through regularization

- overfitting (high-variance) vs underfitting (high-bias). A good way of finding a good bias-variance trade-off is to tune the complexity of the model via regularization, a useful mothod to handle collinearity (high correlation amongst features), filter out noise frojm data and eventually to prevent overfitting.

- concept behind regularization is to introduce additional info (noise) to penalize extreme parameter weights. The most common form of regularization is **L2** regularization and in order to apply it , we need to add the regularization term to the cost function that we defined for logistic regression to shrink the weights.

- via the regularization parameter y, we can then control how well we fit the training data while keeping the the weights small.

- So the C parameter we see in LogisticRegression() comes from a convetion in SVM. C is directly related to the regularization parameter y, which is it inverse.! 

So decreasing the value of the inverse regularization parameter C means we are INCREASING the regularization strength that we can visualize by plotting the L2 regularization path for the 2 weigth coefficients:


```python
weights, params = [], []
for c in np.arange(-5, 5):
    lr = LogisticRegression(C=10**c, random_state=0)
    lr.fit(X_train_std, y_train)
    weights.append(lr.coef_[1])
    params.append(10**c)

weights = np.array(weights)
plt.plot(params, weights[:, 0], label='petal length')
plt.plot(params, weights[:, 1], linestyle='--', label='petal width')
plt.ylabel('weight coefficient')
plt.xlabel('C')
plt.legend(loc='upper left')
plt.xscale('log')
```

we fitted 10 LogisticRegression models with different values  for the inverse-regularization parameter C. For the purpose of illustration, we only collected the weight coefficients of the class 2 vs all classifier (OvR technique used for multi-class classification)

## Maximum margin classification with Support Vector Machines (SVM)

our optimization objective is to maximize the margin (distance between the separating hyperplane and the training samples that are closest to this plane (called support vectors). 

The rationale behind having decision boundaries with large margins is that they tend to have lower generalization error whereas models with small margins are more prone to overfitting.

```python
from sklearn.svm import SVC
svm = SVC(kernel='linear', C=1.0,random_state=0)
svm.fit(X_train_std, y_train)
plot_decision_regions(X_combined_std, y_combined, classifier=svm, test_idx=range(105,150))
plt.xlabel('petal length (standardized)') ; plt.ylabel('petal width (standardized)')
plt.legend(loc = 'upper left')
```

```python
## solving non-linear classification problems. Let's create random noise in a dataset:
np.random.seed(0)
X_xor = np.random.randn(200, 2)
y_xor = np.logical_xor(X_xor[:, 0] > 0,
                       X_xor[:, 1] > 0)
y_xor = np.where(y_xor, 1, -1)

plt.scatter(X_xor[y_xor == 1, 0],
            X_xor[y_xor == 1, 1],
            c='b', marker='x',
            label='1')
plt.scatter(X_xor[y_xor == -1, 0],
            X_xor[y_xor == -1, 1],
            c='r',
            marker='s',
            label='-1')
```

The basic idea behing KERNEL methods to deal with such linearly inseparable data is to create nonlinear combinations of the orginal features and project them onto a higher dimensional space via a mapping function where it becomes linearly separable.

A problem with the kernel approach is that it is computaionally very expensive, espcially with high-dimensional data. This is very the so-called KERNEL-TRICK comes into play. Roughly speaking, the term kernel can be interpreted as a similarity function between a pair of samples. The minus sign inverts the distance meausre into a similarity score and , due to the exponential term, the resulting similarity will fall into a range between 1 (for axactly similar samples) and 0 (for very dissimilar samples) 

```python
# let's replace the parameter kernel with 'rbf' (Radial basis function)
svm = SVC(kernel='rbf', random_state=0, gamma=.2, C=10) 
# y parameter gamma: cut-off parameter for the Gaussian sphere. If we increase the y-value, we increase the influence
# of the training samples, leading to softer decision boundaries.
svm.fit(X_xor, y_xor)
plot_decision_regions(X_xor, y_xor, classifier=svm)
plt.legend(loc = 'upper left')
# magic:!!
```

```python
# let's apply the rbf kernel to the Iris dataset. 
svm = SVC(kernel='rbf', random_state=0, gamma = 40, C=1.0)
svm.fit(X_train_std, y_train)
plot_decision_regions(X_combined_std, y_combined, classifier=svm, test_idx=range(105, 150))
# This is a too good of a fit. Controlling y-parameter is important in controlling overfitting.
```

## Decision tree learning

split the data on the feature that results in the highest IG (Information Gain). In an iterative process, we can keep on splitting at each child-node until the leaves are pure. This can result into a very deep tree with manu nodes, whcih can easily lead to overfitting. Thus we typically want to prune the tree by setting a limit for the maximal depth of the tree. 

- IG is the difference between the impurity of the parent node and the sum of the child node impurities. - the lower the impurity of the child node, the higher the IG.

- 3 most commonly used splitting criteria in binary decision trees are **GINI IMPURITY**, **ENTROPY** and **CLASSIFIFCATION ERROR** 

- The entropy criterion attempts to maximize the mutual IG, the GINI impurity can be understood as a criterion to minimize the probability of a misclassification.

- in practise both Gini impurity and entropy yield similar results, it's more worthwile spending time experimenting on different **pruning cut-offs**

```python
# let`s plot the different impurity indices
def gini(p):
    return p * (1 - p) + (1 - p) * (1 - (1 - p))


def entropy(p):
    return - p * np.log2(p) - (1 - p) * np.log2((1 - p))


def error(p):
    return 1 - np.max([p, 1 - p])

x = np.arange(0.0, 1.0, 0.01)

ent = [entropy(p) if p != 0 else None for p in x]
sc_ent = [e * 0.5 if e else None for e in ent]
err = [error(i) for i in x]

fig = plt.figure()
ax = plt.subplot(111)
for i, lab, ls, c, in zip([ent, sc_ent, gini(x), err], 
                          ['Entropy', 'Entropy (scaled)', 
                           'Gini Impurity', 'Misclassification Error'],
                          ['-', '-', '--', '-.'],
                          ['black', 'lightgray', 'red', 'green', 'cyan']):
    line = ax.plot(x, i, label=lab, linestyle=ls, lw=2, color=c)

ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15),
          ncol=3, fancybox=True, shadow=False)

ax.axhline(y=0.5, linewidth=1, color='k', linestyle='--')
ax.axhline(y=1.0, linewidth=1, color='k', linestyle='--')
plt.ylim([0, 1.1])
plt.xlabel('p(i=1)')
plt.ylabel('Impurity Index')
plt.tight_layout()
```

## combining weak to strong learners via Random Forests

RF is an ensemble of decision trees. Ensemble models combine **weak learners** to build a more robust model, a **strong learner** that has a better generalization error and is less susceptible to overfitting.

steps:
- draw random bootstrap sample of size n (randomly choose n samples from the training set with replacement.
- grow a decision tree. At each node: randomly select d features w/o replacement AND split the node using the features that provide the best split according to the objective function
- repeat 1 + 2 k times
- aggregate the prediction by each tree to assign the class label by majority vote. 

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(criterion= 'entropy',
                                n_estimators = 10, 
                                random_state=0,
                                n_jobs=2)

forest.fit(X_train, y_train)
plot_decision_regions(X_combined, y_combined, classifier=forest, test_idx=range(105,150))
```

## K_nearest neighbors - lazy learning algorithm

Called lazy because it doesn't learn a discriminative function from the training data but memorizes the training data instead. 

- choose number of k and a distance metric
- find the k nearest neigborsof the sample that we want to classify
- assign the class label by majority vote

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5, p=2, metric= 'minkowski')
knn.fit(X_train_std, y_train)
plot_decision_regions(X_combined_std, y_combined, classifier=knn, test_idx=range(105, 150))

plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
```

Note that the distance metric will have impact on the result s and that some distance metrics require scaling like the **Euclidean distance**. The **minkowski** metric is a generalization of the Euclidean and Manhattan distance
