### Ordinary Least Squares regression model (OLS)

OLS minimizes the sum of the squared vertical distances (residuals/ errors) to the sample points. The cost function in OLS is idnetical to the one ADALINE (**Sum of Squared Errors - SSE**) but w/o the unit step function so that we obtain continuous variables instead a class labels between -1 and 1. To demonstrate its similarity, let`s take the GD (**Gradient Descent**) implemenatation from **Adaline** and remove the unit step function:

```python
class LinearRegressionGD(object):
    
    """
    Parameters
    -----------
    eta : float
        learning rate (between 0.0 and 1.0)
    n_iter : int
        passes over the training set
    
    Attributes
    ----------
    w_ : 1-D array
        Weights after fitting
    
    """

    def __init__(self, eta=0.001, n_iter=20):
        self.eta = eta
        self.n_iter = n_iter

    def fit(self, X, y):
        
        """Fit training data.

        Parameters
        ----------
        X : {array-like}, shape = [n_samples, n_features]
            Training vectors, where n_samples is the number of samples and
            n_features is the number of features.
        y : array-like, shape = [n_samples]
            Target values.

        Returns
        -------
        self : object

        """
        self.w_ = np.zeros(1 + X.shape[1])
        self.cost_ = []
        
        for i in range(self.n_iter):
            output = self.net_input(X)
            errors = (y - output)
            self.w_[1:] += self.eta * X.T.dot(errors)
            self.w_[0] += self.eta * errors.sum()
            cost = (errors**2).sum() / 2.0
            self.cost_.append(cost)
        return self

    def net_input(self, X):
        """Calculate net input"""
        return np.dot(X, self.w_[1:]) + self.w_[0]

    def predict(self, X):
        return self.net_input(X)
```

```python
X = df[['RM']].values
y = df['MEDV'].values
```

```python
# standardize the variables for better convergence of the GD algorithm
sc_x = StandardScaler()
sc_y = StandardScaler()
X_std = sc_x.fit_transform(X)
# Just passing:  y_std = sc_y.fit_transform(y) will throw warning: passing 1d arrays as data is deprecated in 
# 0.17 and will raise ValueError in 0.19. Reshape your data either using X.reshape(-1, 1) if your data has a 
# single feature or X.reshape(1, -1) if it contains a single sample. So:
y_std = sc_y.fit_transform(y[:, np.newaxis]).flatten()

lr = LinearRegressionGD()
lr.fit(X_std, y_std)
```

```python
# Good idea to plot the cost of the function against the epochs (passes over the training set) if we are using 
# optimization algorithms. We'll see that the GD converges after the 5th epoch
plt.plot(range(1, lr.n_iter + 1), lr.cost_)
plt.ylabel('SSE'); plt.xlabel('Epochs')
```

```python
# how well does the LR model fit the data. Let's plot with a helper function:
def line_regplot(X, y, model):
    plt.scatter(X ,y, c = 'green', alpha = .6)
    plt.plot(X, model.predict(X), color= 'red', ls = '--')
    return None

line_regplot(X_std, y_std, lr)
plt.xlabel('Average number of rooms [RM] (standardized)')
plt.ylabel('Price in $1000\'s [MEDV] (standardized)')
```

We don't have to update the weights of the intercept if we are working with standardized data since the y_intercept is always 0. Let`s confirm this by printing the weights:

```python
print('Slope: %.3f' % lr.w_[1])
print('Intercept: %.3f' % lr.w_[0])
```

```python
# to scale the predicted price outcome back to its original scale:
num_rooms_std = sc_x.transform(np.array([[5.0]])) # house with 5 rooms
price_std = lr.predict(num_rooms_std)
print("Price in $1000's: %.3f" % sc_y.inverse_transform(price_std))
```

### Fitting a robust regression model using RANSAC

As an alternative to throwing out **outliers** we can use a robust method of regression called **RANdom SAmple Concensus (RANSAC) ** that fit a regression model to a subset of the data , the so-called **inliers**. !

- select a random number of samples to be inliers
- test all other data points against the fitted model and those points that fall within a user-defined tolerance to the inliers
- refit using all inliers
- estimate the error of the fitted model against the inliers
- Terminate if the model meets a certain user-defined threshold or if a fixed number of iterations has been reached

```python
from sklearn.linear_model import RANSACRegressor
ransac = RANSACRegressor(LinearRegression(), 
                        max_trials = 100,
                        min_samples = 50, # set of randomly chosen samples 
                        residual_metric = lambda x: np.sum(np.abs(x), axis= 1), 
                        residual_threshold = 5.0,
                         # only allow samples to be included in the inliers if their vertical distance 
                         # to the fitted line is with 5 distance points
                        random_state = 0)
ransac.fit(X, y)
```

Normally sklearn would use MAD (**Median Absolute Deviation**) estimate to select the inlier threshold but there are many new methods to detect such a threshol automatically.

```python
inlier_mask = ransac.inlier_mask_
outlier_mask = np.logical_not(inlier_mask)  # np.logical_not :  Compute the truth value of NOT x element-wise.

line_X = np.arange(3, 10, 1)
line_y_ransac = ransac.predict(line_X[:, np.newaxis])

plt.scatter(X[inlier_mask], y[inlier_mask],
            c='blue', marker='o', label='Inliers')

plt.scatter(X[outlier_mask], y[outlier_mask],
            c='lightgreen', marker='s', label='Outliers')

plt.plot(line_X, line_y_ransac, color='red')   
plt.xlabel('Average number of rooms [RM]'); plt.ylabel('Price in $1000\'s [MEDV]')
plt.legend(loc='upper left')
```

```python
print ('Slope: %.3f' % ransac.estimator_.coef_[0]) # steeper slope as compared to the non-Ransac approach (9.2)
print ('Intercept: %.3f' % ransac.estimator_.intercept_)
```

### Evaluating the performance of linear regression models

```python
from sklearn.cross_validation import train_test_split
X = df.iloc[:, :-1].values # use all variables to train a multiple regression model
y = df['MEDV'].values

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=0)

slr = LinearRegression()

slr.fit(X_train, y_train)
y_train_pred = slr.predict(X_train)
y_test_pred = slr.predict(X_test)
```

Since our model has multiple explanatory variables we cannot visualize the linear regression line(or hyperplane) in a 2-D plot but we can plot the residuals (differences between actual and predicted values) vs the predicted values. Those **residual plots** are commonly used to detect nonlinearity and outliers and to check if the errors are randomly distributed.

If we see patterns in a residual plot it means that the model is unable to capture some explanatory information, which is leaked into the residuals as we can slightly see in the plot below:

```python
plt.scatter(y_train_pred, y_train_pred - y_train, c= 'blue', alpha = .8, s = 25, label = 'training data')
plt.scatter(y_test_pred, y_test_pred - y_test, c= 'green', alpha = .8, s = 25, label = 'test data')
plt.hlines(y=0, xmin=-10, xmax=50, lw=2, color='red')
plt.xlim([-10, 50])
plt.legend(loc = 'upper left')
```

```python
# calculate the Mean Squared Error (MSE):
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

print('MSE train: %.3f, test: %.3f' % (
        mean_squared_error(y_train, y_train_pred),
        mean_squared_error(y_test, y_test_pred)))
print('R^2 train: %.3f, test: %.3f' % (
        r2_score(y_train, y_train_pred),
        r2_score(y_test, y_test_pred)))
```

MSE is larger on the test set meaning we are overfitting. 
The **R squared, or the coefficient of determination** can be seen as a standardized version of the MSE, for better interpretability. 

### Using regularized methods for regression

most popular approaches to regularized regression techniques:
- Ridge --> L2- penalized model where we simply add the squared sum of the weights to out least-squares cost function
- LASSO **Least Absolute Shrinkage and Selection Operator** --> depeneding on the regularization strength, certain weights can become 0 and therefor makes it also useful as a supervised selection technique. A downside of LASSO is that we can only select n variables if m > n
- elastic Net --> has a L1 penalty to create sparsity and a L2 penalty to overcome some of the limitations of the LASSO, such as the number of variables. 

```python
from sklearn.linear_model import Lasso, Ridge, ElasticNet
# plug in a Lasso variant of the model
lasso = Lasso(alpha=0.1)
lasso.fit(X_train, y_train)
y_train_pred = lasso.predict(X_train)
y_test_pred = lasso.predict(X_test)
print(lasso.coef_)
```

```python
print('MSE train: %.3f, test: %.3f' % (
        mean_squared_error(y_train, y_train_pred),
        mean_squared_error(y_test, y_test_pred)))
print('R^2 train: %.3f, test: %.3f' % (
        r2_score(y_train, y_train_pred),
        r2_score(y_test, y_test_pred)))
```

```python
# plug in a Ridge variant of the model
ridge = Ridge(alpha=1)
ridge.fit(X_train, y_train)
y_train_pred = ridge.predict(X_train)
y_test_pred = ridge.predict(X_test)
print(ridge.coef_)
```

```python
print('MSE train: %.3f, test: %.3f' % (
        mean_squared_error(y_train, y_train_pred),
        mean_squared_error(y_test, y_test_pred)))
print('R^2 train: %.3f, test: %.3f' % (
        r2_score(y_train, y_train_pred),
        r2_score(y_test, y_test_pred)))
```

### Turning a linear regression model into a curve - polynomial regression

In previsous examples we assumed a linear relationship between the variables. One way to account for the violation of linearity assumptions is to use a polynomial regression model by adding polynomial terms:

y = W0 + W1.X + W2.X2X2

implement the PolynomialFeature transformer from sklearn to add a quadratic term (d= 2) to a simple regression problem with one explanatory variable, and compare the polynomial fit to the linear fit:

```python
X = np.array([258.0, 270.0, 294.0, 
              320.0, 342.0, 368.0, 
              396.0, 446.0, 480.0, 586.0])[:, np.newaxis]

y = np.array([236.4, 234.4, 252.8, 
              298.6, 314.2, 342.2, 
              360.8, 368.0, 391.2,
              390.8])
```

```python
## add a second degree Polynomial term
from sklearn.preprocessing import PolynomialFeatures # Generate polynomial and interaction features.
lr = LinearRegression()
pr = LinearRegression()
quadratic = PolynomialFeatures(degree=4)
X_quad = quadratic.fit_transform(X)
```

```python
from IPython.display import display
display (X)
display (X_quad)
```

```python
# fit a simple regression model for camparison
# fit linear features
lr.fit(X, y)
X_fit = np.arange(250, 600, 10)[:, np.newaxis]
y_lin_fit = lr.predict(X_fit)

# fit quadratic features
pr.fit(X_quad, y)
y_quad_fit = pr.predict(quadratic.fit_transform(X_fit))

# plot results
plt.scatter(X, y, label='training points')
plt.plot(X_fit, y_lin_fit, label='linear fit', linestyle='--')
plt.plot(X_fit, y_quad_fit, label='quadratic fit')
plt.legend(loc='upper left')
```

```python
# compare the MSE's:
y_lin_pred = lr.predict(X)
y_quad_pred = pr.predict(X_quad)

print('Training MSE linear: %.3f, quadratic: %.3f' % (
        mean_squared_error(y, y_lin_pred),
        mean_squared_error(y, y_quad_pred)))
print('Training R^2 linear: %.3f, quadratic: %.3f' % (
        r2_score(y, y_lin_pred),
        r2_score(y, y_quad_pred)))
```

### Modeling nonlinear relationships in the Housing Dataset

model the leationship between the houseprices and LSTAT (% of lower status of the population) using second degree (quadratic) and third degree (cubic polynomials and compare to the linear fit:

```python
X = df[['LSTAT']].values
y = df['MEDV'].values
regr = LinearRegression()

# create polynomial features
quadratic = PolynomialFeatures(degree = 2)
cubic = PolynomialFeatures(degree= 3)
X_quad = quadratic.fit_transform(X)
X_cubic = cubic.fit_transform(X)

# fit features
X_fit = np.arange(X.min(), X.max(), 1)[:, np.newaxis]

regr = regr.fit(X, y) # linear fit
y_lin_fit = regr.predict(X_fit)
linear_r2 = r2_score(y, regr.predict(X))

regr = regr.fit(X_quad, y) # quadratic fit
y_quad_fit = regr.predict(quadratic.fit_transform(X_fit))
quadratic_r2 = r2_score(y, regr.predict(X_quad))

regr = regr.fit(X_cubic, y) # cubic fit
y_cubic_fit = regr.predict(cubic.fit_transform(X_fit))
cubic_r2 = r2_score(y, regr.predict(X_cubic))

# plot results
plt.scatter(X, y, label='training points', color='lightgray')

plt.plot(X_fit, y_lin_fit, 
         label='linear (d=1), $R^2=%.2f$' % linear_r2, 
         color='blue', 
         lw=2, 
         linestyle=':')

plt.plot(X_fit, y_quad_fit, 
         label='quadratic (d=2), $R^2=%.2f$' % quadratic_r2,
         color='red', 
         lw=2,
         linestyle='-')

plt.plot(X_fit, y_cubic_fit, 
         label='cubic (d=3), $R^2=%.2f$' % cubic_r2,
         color='green', 
         lw=2, 
         linestyle='--')

plt.xlabel('% lower status of the population [LSTAT]')
plt.ylabel('Price in $1000\'s [MEDV]')
plt.legend(loc='upper right')
```

Polynomial fits increases the chance of overfitting so off course always evaluate the generalization performance on a separate test set. In addition, Polynomial features are not always the best choice for modeling non-linear relationships. Fi looking at the MEDV-LSTAT scatterplot, we could propose a **log transfirm** of the LSTAT feature variable and the square root of MEDV may project the data onto a linear feature subspace suitable for linear regression fit. Let's test the hypothesis:

```python
# transform features
X_log = np.log(X)
y_sqrt = np.sqrt(y)

# fit features
X_fit = np.arange(X_log.min()-1, X_log.max()+ 1, 1)[: ,np.newaxis]
regr = regr.fit(X_log, y_sqrt)
y_lin_fit = regr.predict(X_fit)
linear_r2 = r2_score(y_sqrt, regr.predict(X_log))

# plot results
plt.scatter(X_log, y_sqrt, label='training points', color='lightgray')

plt.plot(X_fit, y_lin_fit, 
         label='linear (d=1), $R^2=%.2f$' % linear_r2, 
         color='blue', 
         lw=2)

plt.xlabel('log(% lower status of the population [LSTAT])')
plt.ylabel('$\sqrt{Price \; in \; \$1000\'s [MEDV]}$')
plt.legend(loc='lower left')
```

After transforming the explanatory variables onto a log subspace and taking the square root of the target variable, we were able to capture the relationship much better than any previous polynomial fit!

### Dealing with nonlinear relationships using random forests

Ensemble of multiple decision trees can be understood as the sum of peice-wise linear functions in contrast to the global linear and polynomial regression models: we are sub-dividing the input space into smaller regions that become much more manageable.

An advantage of random forest regressors (and decision tree regressors in general) is that we don't have to transform our features if we are dealing with nonlinear data. We grow our tree by iteratively splitting its nodes until the leaves are pure or a stopping criterion is satisfied. 

To use a tree for regression we will replace the **entropy** as the impurity measure by the MSE: in the context of decision tree regression, the MSE is often referred to as the **within-node variance**, which is why splitting criterion is also known as variance reduction. 

```python
def line_regplot(X, y, model):
    plt.scatter(X ,y, c = 'green', alpha = .6)
    plt.plot(X, model.predict(X), color= 'red', ls = '--')
    return None
```

```python
from sklearn.tree import DecisionTreeRegressor
X = df[['LSTAT']].values
y = df['MEDV'].values

tree = DecisionTreeRegressor(max_depth=3)
tree.fit(X, y)

sort_idx = X.flatten().argsort()

line_regplot(X[sort_idx], y[sort_idx], tree)
plt.xlabel('% lower status of the population [LSTAT]')
plt.ylabel('Price in $1000\'s [MEDV]')
```

** Random forest regression** often have a better generalization performance due to the randomness that helps to reduce the model variance and they are less sensitive to outliers in the data and don't require much parameter tuning.

The only parameter we typically tune is the number of trees in the ensemble. The RandomForestRegressor() is almost equal to RandomForestClassifier() witht he only difference that we use the MSE criterion to grow the individual decision trees, and the predicted target variable is calculated as the avrage prediction over all decision trees.

```python
from sklearn.ensemble import RandomForestRegressor

X = df.iloc[:, :-1].values
y = df['MEDV'].values

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.4, random_state=1)


forest = RandomForestRegressor(n_estimators=1000, 
                               criterion='mse', 
                               random_state=1, 
                               n_jobs=-1)
forest.fit(X_train, y_train)
y_train_pred = forest.predict(X_train)
y_test_pred = forest.predict(X_test)

print('MSE train: %.3f, test: %.3f' % (
        mean_squared_error(y_train, y_train_pred),
        mean_squared_error(y_test, y_test_pred)))
print('R^2 train: %.3f, test: %.3f' % (
        r2_score(y_train, y_train_pred),
        r2_score(y_test, y_test_pred)))
```

```python
plt.scatter(y_train_pred,  
            y_train_pred - y_train, 
            c='black', 
            marker='o', 
            s=35,
            alpha=0.5,
            label='Training data')
plt.scatter(y_test_pred,  
            y_test_pred - y_test, 
            c='lightgreen', 
            marker='s', 
            s=35,
            alpha=0.7,
            label='Test data')

plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.legend(loc='upper left')
plt.hlines(y=0, xmin=-10, xmax=50, lw=2, color='red')
plt.xlim([-10, 50])
```
