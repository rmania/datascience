```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
# start with a 2 dim dataset for visualization purposes
X, y = make_blobs(n_samples=150, 
                  n_features=2, 
                  centers=3, 
                  cluster_std=0.4, 
                  shuffle=True, 
                  random_state=0)
```

```python
plt.scatter(X[:,0], X[:,1], s= 50, c='blue')
plt.grid()
```

Four k-Means steps:

- randomly pick k centroids from the sample points as initial cluster centers
- Assign each sample to the nearest centroid
- Move the centroids to the center of the samples that were assigned to it
- repeat steps 2/3 until cluster assignment do not change or a user-defined tolerance or a maximum number of iterations is reached.

A commonly used distance metric is the **squared Euclidean distance** between 2 points. BAsed on this distance metric we can describe the k-means algorithm as an optimization problem, an iterative approach for minimizing the **within-cluster sum of squared errors (SSE)**

```python
km.cluster_centers_
```

```python
km = KMeans(n_clusters=3, 
            init='random', 
            n_init=10, 
            max_iter=300,
            tol=1e-04, # Relative tolerance with regards to inertia to declare convergence
            random_state=0)
y_km = km.fit_predict(X)

plt.scatter(X[y_km == 0, 0],
            X[y_km == 0, 1],
            s=50, c='lightgreen', marker='s', label='cluster 1')
plt.scatter(X[y_km == 1, 0],
            X[y_km == 1, 1],
            s=50, c='orange', marker='o', label='cluster 2')
plt.scatter(X[y_km == 2, 0],
            X[y_km == 2, 1],
            s=50, c='lightblue', marker='v', label='cluster 3')
plt.scatter(km.cluster_centers_[:, 0],
            km.cluster_centers_[:, 1],
            s=250, marker='*', c='red', label='centroids')
plt.legend()
plt.grid()
plt.tight_layout()
```

```python
# we dont need to compute the within-cluster SSE explicitly as it is already accessible through the intertia_ attribute:
# Inertia: sum of distances of samples to their closest cluster center.
print (' Distortion: %.2f' % km.inertia_)
```

### K-means ++ 

In classic k-means clustering the centroids are chosen randomly and if choosen poorly can take a while to converge or result in bad clusters. On way to address this is to run KM multiple times and choose the best performing model in terms of SSE. Another strategy is to choose the initial centroids far away from each other via the **K means++ algorithm** which leads to better and more consistent results.

### Hard vs soft clustering

Hard clsutering assigns each sample to exactly one cluster, like KM. ** Soft Clustering or fuzzy clustering** assign a sample to one or more clusters. **Fuzzy C-means (FCM), or Soft k-means** Here we replace the hard cluster assignment by probabilities for each points belonging to each cluster.

### using the elbow method to find the optimal number of clusters

Based on the within-cluster SSE, we can use a graphical tool, the so-called **elbow method**, to estimate the optimal number of clusters. Intuitively, we can say that if K increases, the distortion decreases. This is because the samples will be closer to the centroids they were assigned to. Elbow method identifies the value of k where the distortion begins to increase most rapidly.

```python
plt.subplots(figsize =[10,5])
distortions = []
for i in range (1, 11):
    km = KMeans(n_clusters=i, init ='k-means++', n_init=10, max_iter=300, random_state=0)
    km.fit(X)
    distortions.append(km.inertia_)
plt.plot(range(1,11), distortions)
plt.xlabel('Number of clusters'); plt.ylabel('Distortion')
```

### Quantifying the quality of clustering via silhouette plots

Another intrinsic metric to evaluate the quality of clustering is the silhouette analysis: can be used to plot a measure of how tightly grouped the samples in the clusters are.

- calculate cluster cohesion as the average distance between a sample x and all other points in the same cluster
- calculate cluster separation from the next closest cluster as the average distance between a sample x and all other points in the same nearest cluster
- calculate the silhouettes the difference between cluster cohesion and cluster separation, divided by the greater one of those 2

```python
from matplotlib import cm
from sklearn.metrics import silhouette_samples

plt.subplots(figsize=[10,6])
km = KMeans(n_clusters=3, 
            init='k-means++', 
            n_init=10, 
            max_iter=300,
            tol=1e-04,
            random_state=0)
y_km = km.fit_predict(X)

cluster_labels = np.unique(y_km)
n_clusters = cluster_labels.shape[0]
silhouette_vals = silhouette_samples(X, y_km, metric='euclidean')

y_ax_lower, y_ax_upper = 0, 0
yticks = []
for i, c in enumerate(cluster_labels):
    c_silhouette_vals = silhouette_vals[y_km == c]
    c_silhouette_vals.sort()
    y_ax_upper += len(c_silhouette_vals)
    color = cm.jet(float(i) / n_clusters)
    plt.barh(range(y_ax_lower, y_ax_upper), c_silhouette_vals, height=1.0, 
             edgecolor='none', color=color)

    yticks.append((y_ax_lower + y_ax_upper) / 2.)
    y_ax_lower += len(c_silhouette_vals)
    
silhouette_avg = np.mean(silhouette_vals)
plt.axvline(silhouette_avg, color="red", linestyle="--") 

plt.yticks(yticks, cluster_labels + 1)
plt.ylabel('Cluster')
plt.xlabel('Silhouette coefficient')

# our clusters are pretty good as indicated by high average silhouette coefficients!
```

### Hierarchical trees

Hierarchical clustering allows for dendogram plotting and doesn't need prespecified number of clusters. The main approaches are **agglomerative and divisive**. In Divisive hierarchical clustering, we start with one cluster that encompasses all of our samples, and we iteratively split the cluster into smaller subparts until each cluster only contains one sample. In the examples beneath we take the agglomerative approach which is the opposite: we start with each sample as an individual cluster and we merge the closest points until only one cluster remains.

The 2 standard algorithms for agglomerative hierarchical clustering are **single linkage** and **complete linkage**. In single linkage, we compute the distances between the most similar members for each pair of clusters and merge the 2 clusters for which the distance between the most similar members is the smallest. The complete linkage approach is the same but instead we take the the most dissimilar members to perform the merge. 

So **agglomerative hierarchical clustering** steps:
- compute distance matrix of all samples
- represent each data point a a singleton cluster
- merge the 2 closest clusters based on the 2 most dissimilar (distant) members *see complete linkage)
- update the similarity matrix
- repeat steps 2-4 until one cluster remains

```python
np.random.seed(123)
variables = ['X', 'Y', 'Z']
labels = ['ID_0', 'ID_1', 'ID_2', 'ID_3', 'ID_4']

X = np.random.random_sample([5, 3])*10
df = pd.DataFrame(X, columns=variables, index=labels)
df
```

```python
# calculate the distance matrix: Euclidean distance between each pair based on the features X,Y, Z
from scipy.spatial.distance import pdist, squareform
row_dist = pd.DataFrame(squareform(
                        pdist(df, metric = 'euclidean')),
                        columns= labels, index= labels)
row_dist

# squareform: Converts a vector-form distance vector to a square-form distance matrix, and v.v.
```

We can either pass a condensed distance matrix (upper triangular) from the pdist function, or we can pass the "original" data array and define the metric='euclidean' argument in linkage. However, we should not pass the squareform distance matrix, which would yield different distance values although the overall clustering could be the same.

```python
# 1. incorrect approach: Squareform distance matrix

from scipy.cluster.hierarchy import linkage

row_clusters = linkage(row_dist, method='complete', metric='euclidean')
pd.DataFrame(row_clusters,
             columns=['row label 1', 'row label 2',
                      'distance', 'no. of items in clust.'],
             index=['cluster %d' % (i + 1)
                    for i in range(row_clusters.shape[0])])
```

```python
# 2. correct approach: Condensed distance matrix

row_clusters = linkage(pdist(df, metric='euclidean'), method='complete')
pd.DataFrame(row_clusters,
             columns=['row label 1', 'row label 2',
                      'distance', 'no. of items in clust.'],
             index=['cluster %d' % (i + 1) 
                    for i in range(row_clusters.shape[0])])
```

```python
# 3. correct approach: Input sample matrix

pd.DataFrame(row_clusters,
             columns=['row label 1', 'row label 2',
                      'distance', 'no. of items in clust.'],
             index=['cluster %d' % (i + 1)
                    for i in range(row_clusters.shape[0])])
```

The linkage matrix consist of several rows where each row represents a merge. The 1st and second columns denote the most dissimilar members in each cluster. 

```python
from scipy.cluster.hierarchy import dendrogram

row_dendr = dendrogram(row_clusters, labels=labels)
plt.tight_layout()
plt.ylabel('Euclidean distance')
```

### Attaching dendrograms to a heat map

```python
""" 
fig.add_axes = Add axes at position *rect* [*left*, *bottom*, *width*, *height*] all quantities are in fractions of figure
So we define an x-axis position, width and height of the dendogram via the add_axes attribute.
We rotate the the dendogram 90 degrees counter-clockwise
"""

# plot row dendrogram
fig = plt.figure(figsize=(8, 8), facecolor='white')
axd = fig.add_axes([0.09, 0.1, 0.2, 0.6])

# note: for matplotlib < v1.5.1, please use orientation='right'
row_dendr = dendrogram(row_clusters, orientation='left')

# reorder data with respect to clustering
df_rowclust = df.ix[row_dendr['leaves'][::-1]]

axd.set_xticks([])
axd.set_yticks([])

# remove axes spines from dendrogram
for i in axd.spines.values():
        i.set_visible(False)

# plot heatmap, position it right next to the dendrogram.
axm = fig.add_axes([0.23, 0.1, 0.6, 0.6])  # x-pos, y-pos, width, height
cax = axm.matshow(df_rowclust, interpolation='nearest', cmap='hot_r')
fig.colorbar(cax)
axm.set_xticklabels([''] + list(df_rowclust.columns)) # (X, y, Z)
axm.set_yticklabels([''] + list(df_rowclust.index)) # ID_i
```

### Locating regions of high density via DBSCAN

At least introduce one more clustering apporach. That of the **Density-based Spatial Clustering of Applications with Noise**. Here a special label is assigned to each sample point based on the following criteria:
- point is considered a **core point** if at least a specified number (MinPts) of neighboring points fall within the specified radius (E)
- A **border point** is a opint that has fewer neighbors than MinPts within E, but lies within the E radius of a core point
- All other points are considered **noise points**

After this classification of points the DBSCAN algorithm conducts 2 steps:
- form a separate cluster for each core point or a connected group of core points (core points are connected if they are no farther away than E)
- Assign each border point to the cluster of its corresponding core point

https://blog.dominodatalab.com/topology-and-density-based-clustering/

```python
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=200, noise= .05, random_state=0)
plt.scatter(X[:,0], X[:, 1])
```

```python
# detect the clusters
f, (ax1, ax2) = plt.subplots(1, 2, figsize= [15,5])
km = KMeans(n_clusters=2, random_state=0)
y_km = km.fit_predict(X)

ax1.scatter(X[y_km == 0, 0], X[y_km == 0, 1], c='lightblue', marker='o', s=40, label='cluster 1')
ax1.scatter(X[y_km == 1, 0], X[y_km == 1, 1], c='red', marker='s', s=40, label='cluster 2')
ax1.set_title('K-means clustering')

ac = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='complete')
y_ac = ac.fit_predict(X)

ax2.scatter(X[y_ac == 0, 0], X[y_ac == 0, 1], c='lightblue', marker='o', s=40, label='cluster 1')
ax2.scatter(X[y_ac == 1, 0], X[y_ac == 1, 1], c='red', marker='s', s=40, label='cluster 2')
ax2.set_title('Agglomerative clustering')

plt.legend()
```

```python
from sklearn.cluster import DBSCAN # Finds core samples of high density and expands clusters from them.

db = DBSCAN(eps=0.2, min_samples=5, metric='euclidean')
y_db = db.fit_predict(X)
plt.scatter(X[y_db == 0, 0], X[y_db == 0, 1], c='lightblue', marker='o', s=40, label='cluster 1')
plt.scatter(X[y_db == 1, 0], X[y_db == 1, 1], c='red', marker='s', s=40, label='cluster 2')
plt.legend()
```

DBSCAN(eps=0.2) --> maximum distance between two samples for them to be considered as in the same neighborhood. We have 2 hyper parameters (MinPts and E radius) that have to be optimized to yield good clustering results.
